```python
import pandas as pd
msg=pd.read_csv('C:\\Users\\exam2\\Downloads\\document.csv',names=['message','label'])
msg
```

```python
print("the dimensions of the dataset are:",msg.shape)
```

```python
msg['labelnum']=msg.label.map({'pos':1,'neg':0})
msg
```

```python
X=msg.message
y=msg.labelnum
```

```python
from sklearn.model_selection import train_test_split
#splitting the dataset into train and test data
xtrain,xtest,ytrain,ytest=train_test_split(X,y)
print(xtrain)
print(ytrain)
print(xtest)
print(ytest)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#output the words or Tokens in the text documents
cv=CountVectorizer()
xtrain_dtm=cv.fit_transform(xtrain)
xtest_dtm=cv.transform(xtest)
print('\n The words or Tokens in the text documents \n')
print(cv.get_feature_names())

df=pd.DataFrame(xtrain_dtm.toarray(),columns=cv.get_feature_names())

```

```python
from sklearn.naive_bayes import MultinomialNB
# Training Naive Bayes (NB) classifier on training data.

clf=MultinomialNB()
clf.fit(xtrain_dtm,ytrain)
predicated=clf.predict(xtest_dtm)
print(predicated)
```

```python
from sklearn import metrics
#printing accuracy, Confusion matrix, Precision and Recall
print("Accuracy:\n",metrics.accuracy_score(ytest,predicated))
print('Recall: ', metrics.recall_score(ytest, predicated))
print('Precision: ', metrics.precision_score(ytest, predicated))
print('Confusion Matrix: \n', metrics.confusion_matrix(ytest, predicated))
```
